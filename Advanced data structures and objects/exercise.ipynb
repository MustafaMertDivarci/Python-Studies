```python
class folder():
    def __init__(self):
        with open("metin.txt","r",encoding="utf-8") as file:
            fileContent=file.read()
            
            words=fileContent.split(" ")
            
            self.simpleWords= list()
            
            
            for i in words:
                i=i.strip(" ")
                i=i.strip(".")
                i=i.strip(",")
                i=i.strip("\n")
                self.simpleWords.append(i)
            
            
                
                
    def allWords(self):
        setWord=set()
        
        for i in self.simpleWords:
            setWord.add(i)
        print("****************************")
        for i in setWord:
            
            print(i)
        print("****************************")
        print("{} Kelime var".format(len(setWord)))
        
        
        
        
    def wordsCount(self):
        wordCounts=dict()
        
        for i in self.simpleWords:
            if (i in wordCounts):
                wordCounts[i]+=1
            else:
                wordCounts[i]=1
        for word,count in wordCounts.items():
            print("Bu kelimeden: {}------{} tane var".format(word,count))
            print("---------------------------------------------------")
        
        
        
        
```
