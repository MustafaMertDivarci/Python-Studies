### Reduce
Gömülü bir fonksiyon\
reduce(function,iterasyon yapılabilen veritipi(liste,demet vs))

```python
from functools import reduce # reduce' yi import ettik
```

```python
def sum(x,y):
    return x+y
```

```python
reduce(sum,[12,18,20,10])
# fonksiyona sırayla x ve y olarak 0. ve 1. indexleri yolluyor
# önce 12,18 i yolladı 30 değerini elde etti.
# [30,20,10] sonra 30 ve 20 yi yollayıp 50 elde edildi.
# [50,10] en son olarak da 50 ve 10 değeri sum fonksiyonuna yollandı.
```

```python
reduce(lambda x,y:x*y,[1,2,3,4,5])
```

```python
def max(x,y):
    if(x>y):
        return x
    else: return y
```

```python
reduce(max,[-2,3,8,2]) 
# sırayla -2 ve 3 ü fonksiyona gönderdi ve fonksiyon 3 ü döndü
# sonra 3 ve 8 i gönderdi fonksiyon 8'i döndü
# son olarak 8 ve 2 yi gönderdi 8 i elde ettik
```
