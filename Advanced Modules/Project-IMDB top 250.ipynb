```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url = "https://www.imdb.com/chart/top/"
```

```python
response=requests.get(url)
```

```python
html_content = response.content
```

```python
print(response)
```

```python
soup = BeautifulSoup(html_content,"html.parser")
```

```python
movieNames=soup.find_all("td",{"class":"titleColumn"})
for i in movieNames: #filmlerin isimlerini aldık
    
    print(i.text)
    print("************************************")
```

```python
movieRatings=soup.find_all("td",{"class":"ratingColumn imdbRating"})

for i in movieRatings:
    
    print(i.text)
    print("******************************************")
```

```python
print((len(movieNames)),(len(movieRatings)))
```

```python
for title,rating in zip(movieNames,movieRatings):
    
    title=title.text
    rating=rating.text
    
    title=title.strip()
    title=title.replace("\n","")
    
    rating=rating.strip()
    rating=rating.replace("\n","")
    
    print("Başlık: ",title)
    print("Rating: ",rating)
```

```python
input_rating=float(input("Higher: "))
```

```python
for title,rating in zip(movieNames,movieRatings):
    
    title=title.text
    rating=rating.text
    
    title=title.strip()
    title=title.replace("\n","")
    
    rating=rating.strip()
    rating=rating.replace("\n","")
    
    if(float(rating)>=input_rating):
        print("Movie Name: {}-----Movie Rating: {}".format(title,rating))
        
    
```
