```python
def calculateGrade(line):
    line=line[:-1]   # sondaki \n i silmiş olduk
    liste=line.split(",")      # liste oluşturup split() metodu ile "," görünce indexledik
    
    name=liste[0]
    midterm1 = int(liste[1])
    midterm2 = int(liste[2])
    final=  int(liste[3])
    
    calculate= midterm1*(0.3) + midterm2 * (0.3) + final*(0.4)
    
    if (calculate>=90):
        letterGrade="AA"
    elif(calculate>=85):
        letterGrade="BA"
    elif(calculate>=80):
        letterGrade="BB"
    elif(calculate>=75):
        letterGrade="CB"
    elif(calculate>=70):
        letterGrade="CC"
    elif(calculate>=65):
        letterGrade="DC"
    elif(calculate>=60):
        letterGrade="DD"
    else: letterGrade="FF"
        
    return name+"-----------------> "+ letterGrade +"\n"
```

```python
with open("dosya.txt","r",encoding="utf-8") as file:   # utf-8 eklemeden çalışmadı.
    addedList=[]
    for i in file:
        addedList.append(calculateGrade(i))
print(addedList)
```

```python
for i in addedList:
    print(i,end="")
```

```python
with open("Grades.txt","w",encoding="utf-8") as file2:
    
    for i in addedList:
        file2.write(i)
    
```
